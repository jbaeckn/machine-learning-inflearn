```python
import pandas as pd
import numpy as np
```

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
         'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
         'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
         'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
         'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}

df = pd.DataFrame(ipl_data)
df
```

```python
df.groupby("Team")["Points"].sum()
```

```python
#list로 반환 가능
df.groupby("Team")["Points"].sum().tolist()
```

```python
df.groupby(['Team', 'Year'])['Points'].sum()
```

```python
#hierarchical index
df.groupby(['Team', 'Year'])['Points'].sum().index
```

```python
h_index = df.groupby(["Team", "Year"])["Points"].sum()
h_index
```

```python
#group으로 묶여진 데이터를 matrix 형태로 전환해줌
h_index.unstack()
```

```python
h_index.swaplevel()
```

```python
#sortlevel로 grouping 가능
h_index.swaplevel().sortlevel(0)
```

# Case Study

```python
df_phone = pd.read_csv('phone_data.csv')
df_phone.head()
```

```python
#change date format
import dateutil
df_phone['date'] = df_phone['date'].apply(dateutil.parser.parse, dayfirst=True)
df_phone.head()
```

```python
#월별로 duration의 합을 보고싶다
df_phone.groupby("month")["duration"].sum()
```

```python
df_phone[df_phone['item'] == 'call'].head()
```

```python
#item이 call이고 network별로 구분하여 duration의 합을 구하기
df_phone[df_phone['item'] == "call"].groupby("network")['duration'].sum()
```

```python
df_phone[df_phone['item'] == "call"].groupby("month")['duration'].sum()
```

```python
#month와 item별로 그룹핑하여 건수를 카운트
df_phone.groupby(['month', 'item'])['date'].count()
```

```python
df_phone.groupby(['month', 'item'])['date'].count().unstack()
```

```python
#month로 그룹핑하여 aggretation을 통해 duration을 합을 구하기
df_phone.groupby('month', as_index=False).agg({'duration':'sum'})
```

```python
#month와 item으로 그룹핑하여 aggregation을 통해 duration의 합, network_type의 수, 각그룹별 date의 첫번째 값을 반환
df_phone.groupby(['month', 'item']).agg({'duration':sum, 'network_type':'count', 'date':'first'})
```

```python
#month와 item으로 구분하여 duration은 min, max, sum, date는 min, 첫번째, unique 값
df_phone.groupby(['month', 'item']).agg({'duration':[min, max, sum], 
                                         'network_type':'count', 
                                         'date':[min, 'first', 'nunique']})
```

```python
grouped = df_phone.groupby('month').agg({'duration':[min, max, np.mean]})
grouped.columns #이 column의 level하나를 없애주고 싶을때
grouped.columns.levels[0]
```

```python
grouped.columns = grouped.columns.droplevel(level=0)
grouped
```

```python
grouped.rename(columns={"min":"min_duration", "max":"max_duration", "mean":"mean_duration"})
#아니면 grouped.add_prefix("duration_") 이렇게 해도 된다
```
